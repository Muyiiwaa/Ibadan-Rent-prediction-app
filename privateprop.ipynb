```python
from bs4 import BeautifulSoup
import requests,re
import csv
import pandas as pd
import numpy as np
```

```python
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36'}

url = 'https://www.privateproperty.com.ng/property-for-rent?search=Ibadan+%2C+Oyo&auto=&ptype=&bedroom=&min_price='
s = requests.Session()
r = s.get(url,timeout=10, headers=headers)
c = r.content
soup = BeautifulSoup(c, 'html.parser')
```

```python
scraped=[]

base_url='https://www.privateproperty.com.ng/property-for-rent?search=Ibadan+%2C+Oyo&auto=&ptype=&bedroom=&min_price=4&page='


for page in range(1,80):
    print(base_url+str(page))
    r= requests.get(base_url+str(page))
    c= r.content
    soup= BeautifulSoup(c,"html.parser")
        
    classes = ["similar-listings-item"]
    for class_ in classes:
        real=soup.find_all("div",{"class":class_})

        for i in list(range(0,len(real))):
            d={}
            d["page" ] = page
            try:
                d["title"]=real[i].find('div', class_='similar-listings-info').get_text().strip()
            except (IndexError,TypeError,AttributeError) as e:
                d["title"]= None
            #try:
                #d["description"]=real[i].find('h4', {'class':'content-title'}).get_text().strip()
            #except (IndexError,TypeError,AttributeError) as e:
                #d["description"]= None
            try:
                d["rooms"] = real[i].find('ul', {'class':'property-benefit'}).get_text().strip()
            except(IndexError,TypeError,AttributeError):
                d['rooms'] = None
            try:
                d["address"] = real[i].find('p', {'class':'listings-location'}).get_text().strip()
            except(IndexError,TypeError,AttributeError):
                d['address'] = None
            try:
                d["price"]=real[i].find('div', {'class':'similar-listings-price'}).get_text().strip()
            except (IndexError,TypeError,AttributeError) as e:
                d["price"] = None            
            scraped.append(d)
            #print(l)
```

```python
df = pd.DataFrame(scraped)
df
```
