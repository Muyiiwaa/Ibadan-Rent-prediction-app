```python
#importing the needed libraries.

import pandas as pd 
import numpy as np
import math
import seaborn as sns
import plotly.express as px
import matplotlib.pyplot as plt
import re
```

```python
df = pd.read_csv('propertypro.csv', index_col = 0)
df = df.drop_duplicates()
```

```python
df.head()
```

```python
def remove_symbols(df, column):
    if column == 'address':
        df[column] = df[column].astype(str).apply(lambda x: re.sub(r'[^\w\s]', '', x))
    else:
        df[column] = df[column].astype(str).apply(lambda x: re.sub(r'[^\d]', '', x))

cols = ['address', 'price']
for col in cols:
    remove_symbols(df, col)
```

```python
df.head()
```

```python
df[['bedroom', 'bathroom','toilet']] = df["rooms"].str.split("\n", n = 2, expand = True)

cols = ['bedroom', 'bathroom','toilet']

for col in cols:
    remove_symbols(df, col)
```

```python
df.head()
```

```python
df['condition'] = df['condition'].fillna('\n')
```

```python
df['condition'].value_counts()
```

```python
df['condition'] = df['condition'].replace({'\n':'Not_new Unfurnished Not_Serviced',
                                           'Newly Built':'Newly_Built Unfurnished Not_Serviced',
                                           'Serviced':'Not_new Unfurnished Serviced',
                                            'Furnished':'Not_new Furnished Not_Serviced',
                                            'Serviced\nNewly Built':'Newly_Built Unfurnished Serviced',
                                            'Furnished\nNewly Built':'Newly_Built Furnished Not_Serviced',
                                            'Furnished\nServiced':'Not_new Furnished Serviced'})

df['condition'] = df['condition'].replace(['Furnished\nServiced\nNewly Built',
                                          'Newly_Built Furnished Serviced'])

# split the condition column

df[['newly_built', 'furnished', 'serviced']] = df['condition'].str.split(' ', 2, expand=True)
```

```python
streets = ['Bodija','Agodi','Oluyole','Iyaganku','Ekotedo','Samonda','Adamasingba','Onikere','Agbowo','Orogun','Odo Ona',
'Apata' 'Aerodrome Gra','Basorun','Eleyele','Okebola','Odinjo','Jericho','Idishin','Ikolaba','Akobo','Ring',
'Academy','challenge','Oke ado','Molete','Olomi','Odo Oba','Ososadi','Inalende','Yemetu','Oje','Oke Aremo','Eleyele',
'Olorunsogo','Ojoo','Agodi','Alakia','Apete','Ariyibi','Akala express','Kolapo Ishola','Sanyo','Moniya','New Bodija','Lagos',
'Onireke Gra','Ojurin','Carlton Gate','Lagelu Estate','Iwo Rd', 'New Garage','Mokola','Apata','Ologuneru','Alalubosa','Soka',
'Heritage Estate','Akala Estate','Ire Akari','Elewuro','Oremeji','Kuola', 'Lagos']

def find_street(column):
    column = column.lower()
    for street in streets:
        if street.lower() in column:
            return street
    return 'other'


df['street'] = df['address'].apply(find_street)
```
